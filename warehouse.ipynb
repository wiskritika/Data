# Data warehouse

what is data warehouse ?
- It is a central location to store every single data .


Data regarding 

- Transcations 
- Sales 
- Budgets 
- Marketing 
- HR 

Database vs Data warehouse 

Database 

- It is used for day-to-day operations 
- Handles current, real time data 
- Supports applications (web apps, mobile apps, ERP, CRM)
- e.g. User login data, orders being placed, payements 

Data warehouse

- Used for analysis and reporting
- Stores historical data from many sources 
- Helps in decision making 
- e.g : Trends over 5 years, monthly revenue reports


Types of data 

Feature ---- database ---- data warehouse
Data ---- current operational ---- historical, analytical 
updates ---- frequent (insert, update, delete) ---- Rare (mostly read -only)
data-source --- usually one application --- multiple database, apis, files 

what is data warehousing?

Process that you follow to fetch the data from database and store it to DATA WARE HOUSE(DWH) with the help of ETL.

ETL (Extract Transform Load)

database --- staging(raw data)---core(curated data) ---datamart(specific-fiance)----datamart(specific-finance)

- Incremental data loading 

staging --- Transient --- Just truncate
staging --- Persistence --- keep the data
