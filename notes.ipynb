git = version control system is a tool that helps to track changes in code 

- it is popular 
- free and open source 
-fast and scalable 

Github = website that allows developer to store and manage their code using git 

- configuring git 

git config --global user.name "Myname"
git config --global user.name "someone@gmail.com"
git config --list

clone = cloning and repository on our local machine 

git clone <-some link->

status = display the state of teh code (git status)

Untracked  = new file that doesn't yet track 
modified = unchanged 



add and commit 

Add = adds new or changes files in your working directory to the git staging area 

git add <--file name-->

Commit = It is the record of change 

git commit -m "some message"



Push command 

push = upload local repo content to remote repo 

git push origin main 
git push -u origin main (get upstream)

init command 

init = used  to create a new git repo 
git init 
git remote add origin <-link->
git remote -v (to verify remote)
git remote -M main(to remove branch)
git push origin main
