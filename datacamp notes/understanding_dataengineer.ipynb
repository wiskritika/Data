# Data structure 

#### Structure Data

- Easy to use 
- consistent model , rows and columns 
- defined types 
- can be grouped to form relation 
- stored in relational database 
- about 20% of the data is structured 
- created and queries using sql 

#### Semi-structure Data

- Easy to search and organize
- consistent model, less rigid implementation : different observations have different sizes 
- different types 
- can be grouped, but needs more work
- NoSQL databases: json,xml,yaml



#### Unstructured Data 

- Does not follow a model , can't be contained in rows and column 
- difficult to search and organize
- usually text, sound, pictures and videos 
- usually stored in data lakes, appears in data warehouses or databases 
- most of the data is unstructured 
- can be extremely valuable 

#### Adding some structure 

- Use AI to search and organize unstructured data 
- Add information to make it semi-structured


# SQL DATABASES

#### SQL 

- Structured query language 
- Industry standard for Realtional Database management system (RDBMS)
- Allows you to access many records at once, and group, filter or aggregate them.
- close to written english, easy to write and understand
- data engineer use SQL to create and maintain databases.
- data scientist use SQL to query (request information from) database 

#### Database schema 

- Databases are made of tables 
- the database schema governs how tables are realted 

#### Data lakes 

- It stores all raw data 
- Can be petabytes (1 million GBs)
- store all data structure 
- cost effective 
- difficult to analyze 
- requires up to date data catalog
- used by data scientists 
- big data, real time analytics 


#### Data warehouse 

- specific data for specific use 
- realtively small 
- stores mainly structured data 
- more costly to update 
- optimized for data analysis 
- also used by data analyst and business analyst 
- ad-hoc, read-only queries 


#### Data catalog for data lakes 

- what is the source of this data ?
- where is the data used ?
- who is the owner of the data ?
- how often is this data updated ?
- good practice in terms of data governance ?
- ensure reproducibility 
- no catalog --> data swamp
- good practice for any data storage solution (Reliabilty, Autonomy, Scability, speed)


#### Database 

- general term 
- loosely defined as organized data stored and accessed on a computer 

#### Data warehouse

- It is type of database
